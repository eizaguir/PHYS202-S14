The Beta(4,3) distribution has density
f(x)={60x3(1−x)200≤x≤1otherwise
The maximum of this distribution is f(0.6)=2.0736 (exactly).
(a) Write a program that uses the Rejection Sampling Method to generate a set of 10,000 samples from this distribution and plot the density distribution, the CDF, and a histogram of the sampled distribution on the same graph.
Note: Do not use the built-in Beta function from numpy.random.

```python
f = lambda x: (60*x**3)*(1-x)**2
x = np.arange(0.,1.,0.001)
plt.plot(x,f(x),lw=3)
plt.ylabel(r'$60x^3(1-x)^2$',fontsize=20)
plt.xlabel(r'$x$', fontsize = 20)
plt.xlim(0.,2.)

plt.show()
```

```python
xmin = 0.
xmax = 1.

subx = x[logical_and((x > xmin),(x < xmax))]
ymax = 2.0736
ymin = 0.

plt.plot(x,f(x),lw=3)
plt.ylabel(r'$60x^3(1-x)^2$',fontsize=20)
plt.xlabel(r'$x$', fontsize = 20)
plt.xlim(0.,2.)

plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--') #vertical left
plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') #vertical right
plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--') #horizontal top

plt.show()

N=10000
samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
samples_y = ymin + (ymax-ymin)*np.random.random_sample(N);
```

```python
#Replot the function
plt.plot(x,f(x),lw=3,label="Density")
plt.ylabel(r'$60x^3(1-x)^2$',fontsize=20)
plt.xlabel(r'$x$', fontsize = 20)
plt.xlim(0.,2.)

plt.plot(samples_x[1:500],samples_y[1:500],'g.')


#plot the box
plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--') #vertical left
plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') #vertical right
plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--') #horizontal top

newmask = (samples_y < f(samples_x))
newmask.sum()

plt.plot(samples_x[1:500],samples_y[1:500],'.')

#Use the mask to show the ones below the function in red
plt.plot(samples_x[newmask[:500]],samples_y[newmask[:500]],'r.')

#plot the box
plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--') #vertical left
plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') #vertical right
plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--') #horizontal top

CDF = lambda x: 10*x**6 - 24*x**5 + 15*x**4
x = np.linspace(0.,1.,1000)

plt.plot(x,CDF(x),color = "Purple",linewidth=5,label="CDF")

plt.hist(samples_x[newmask],80,normed=True,color="Yellow",label="Bins")
plt.xlim(0.,1.5)
plt.legend()
#plt.savefig("MCIntegration.png")
plt.show()
```
