For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.
(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
th0 = 0*np.pi/180 # 0 degrees
t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, expressed in radians

#pass the function deriv to the solver with initial values and integration variable
theta = odeint(deriv,theta_init,t) 

plt.figure(figsize=(10,8),dpi=100)
plt.plot(t,theta[:,0],label="Numerical solution, Theta=0")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)

th0 = 45*np.pi/180 # 45 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=45",color="Red")

th0 = 90*np.pi/180 # 90 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=90",color="Purple")

th0 = 135*np.pi/180 # 135 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=135",color="Green")

th0 = 180*np.pi/180 # 180 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=180",color="Black")


plt.ylim(-5,5)
plt.grid()
plt.legend(loc="best")
plt.show()
```

As the initial angle increases, the Period of the pendulum increases (between pi/4 and 3pi/4).
At intial angel = 0 or pi, the period essentially becomes infinity.

(b) Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for θ0 = 0.95π. Look carefully at their behavior. What is going on in these figures?)

```python
th0 = .95*np.pi # .95pi
t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, expressed in radians

#pass the function deriv to the solver with initial values and integration variable
theta = odeint(deriv,theta_init,t) 

#plt.figure(figsize=(10,8),dpi=100)
plt.plot(t,theta[:,0],label="Theta vs. Time, Theta=.95pi")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)

plt.ylim(-5,5)
plt.grid()
plt.legend(loc="best")
plt.show()
```

This is just the simple plot we've always had, showing how the angle changes with time.

```python
plt.plot(t,theta[:,1],label="Omega vs. Time, Theta=.95pi")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\omega (rad/sec)$',fontsize=20)

plt.ylim(-10,10)
plt.grid()
plt.legend(loc="best")
plt.show()
```

This plot show how the angular velocity (omega) changes with time. It is periodic, showing that at
first the velocity is negative, before reaching the end and swinging the other way, reaching a max
of ~9 rad/sec, then going back down.

```python
plt.plot(theta[:,0],theta[:,1],label="Omega vs. Theta, Theta=.95pi")
plt.xlabel(r'$\theta (rad)$',fontsize=20)
plt.ylabel(r'$\omega (rad/sec)$',fontsize=20)

plt.ylim(-10,10)
plt.grid()
plt.legend(loc="best")
plt.show()
```

This plot shows the phase plot, relating how omega moves with theta. As theta goes from -3 to 0 rad,
the pendulum is increasing speed. From theta = 0 to 3 rad, the pendulum loses speed, and this continues in a cycle.

(c) For the θ0 = 0.95π case, check that energy is conserved by plotting E/m vs. t (where E=mgℓcos(θ)+0.5m(ωr)2).
Is the energy a constant? If not, then the computational errors are producing a non-physical loss (or gain) of energy. Review the documentation for odeint and change the error tolerance value to a tighter number.

```python
E = g*ell*np.cos(theta[:,0]) + .5*(theta[:,1]*ell)**2
plt.plot(t,E,label="E/m vs. Time, Theta=.95pi")
plt.xlabel(r'$t (sec)$',fontsize=20)
plt.ylabel(r'$E/m (met/sec^2)$',fontsize=20)

plt.ylim(-10,20)
plt.grid()
plt.legend(loc="best")
plt.show()
```

```python
#new, tighter error
theta = odeint(deriv,theta_init,t,rtol=30) 
E = g*ell*np.cos(theta[:,0]) + .5*(theta[:,1]*ell)**2
plt.plot(t,E,label="E/m vs. Time, Theta=.95pi")
plt.xlabel(r'$t (sec)$',fontsize=20)
plt.ylabel(r'$E/m (met/sec^2)$',fontsize=20)

plt.ylim(-10,20)
plt.grid()
plt.legend(loc="best")
plt.show()
```
