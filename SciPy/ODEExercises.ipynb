(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
th0 = 0*np.pi/180 # 0 degrees
t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, expressed in radians

#pass the function deriv to the solver with initial values and integration variable
theta = odeint(deriv,theta_init,t) 

plt.figure(figsize=(10,8),dpi=100)
plt.plot(t,theta[:,0],label="Numerical solution, Theta=0")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)

th0 = 45*np.pi/180 # 45 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=45",color="Red")

th0 = 90*np.pi/180 # 90 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=90",color="Purple")

th0 = 135*np.pi/180 # 135 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=135",color="Green")

th0 = 180*np.pi/180 # 180 degrees
theta_init = np.array([th0, omega0])
theta = odeint(deriv,theta_init,t) 
plt.plot(t,theta[:,0],label="Numerical solution, Theta=180",color="Black")


plt.ylim(-5,5)
plt.grid()
plt.legend(loc="best")
plt.show()
```

As the initial angle increases, the Period of the pendulum increases (between pi/4 and 3pi/4).
At intial angel = 0 or pi, the period essentially becomes infinity.

(a) Use Euler's method to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, that is y(1), using a first a step size of 0.25, and then 0.02.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

def EulerMethod(step):
    xmax = 1
    dx = step
    nsteps = int(xmax/dx)+1
    x = np.zeros(nsteps)
    yprime = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    x[0] = 0.
    yprime[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        yprime[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * (x[i]+2*y[i])
        
    return x,y,yprime
```


(b) Plot the results from part (a) for the two different step sizes with the exact solution
y=0.25e2x−0.5x−0.25
on one plot. Include a legend. Which step size gives a more accurate result?

```python
x,y,yprime = EulerMethod(.25)
print "Value of Y when X = 1 is %s with step size of .25:"%y[-1]
print y
x,y,yprime = EulerMethod(.02)
print "Value of Y when X = 1 is %s with step size = .02:"%y[-1]
print y

x,y,yprime = EulerMethod(.25)
plt.plot(x,y,label="Step Size of .25")
x,y,yprime = EulerMethod(.02)
plt.plot(x,y,label="Step Size of .02")
y=0.25*np.e**(2*x)-0.5*x-0.25
plt.plot(x,y,label="Exact")
plt.legend(loc="upper left")
plt.xlim(0,1)

# A Step size of ~.02 gives you the most accurate result 
```

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?


```python
def EulerCromerMethod(step):
    xmax = 1 
    dx = step 
    nsteps = int(xmax/dx)+1
    x = np.zeros(nsteps)
    yprime = np.zeros(nsteps)
    y = np.zeros(nsteps)

    x[0] = 0.
    yprime[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        yprime[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * (x[i+1]+2*y[i])
        
    return x,y,yprime

def MidpointMethod(step):
    xmax = 1 
    dx = step 
    nsteps = int(xmax/dx)+1
    x = np.zeros(nsteps)
    yprime = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    x[0] = 0.
    yprime[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        yprime[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * ((1./2)*(x[i+1]+x[i])+2*y[i])
        
    return x,y,yprime
```

```python
x,y,yprime = EulerCromerMethod(.25)
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error1 = abs(y-yex)

plt.plot(x,y,label="Step Size = .25")
x,y,yprime = EulerCromerMethod(.02)
plt.plot(x,y,label="Step Size = .02")
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error2 = abs(y-yex)
plt.plot(x,yex,label="Exact")
plt.legend(loc="upper left")
plt.xlim(0,1)
plt.show()

x,y,yprime = MidpointMethod(.25)
plt.plot(x,y,label="Step Size = .25")
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error3 = abs(y-yex)

x,y,yprime = MidpointMethod(.02)
plt.plot(x,y,label="Step Size = .02")
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error4 = abs(y-yex)

plt.plot(x,y,label="Exact")
plt.legend(loc="upper left")
plt.xlim(0,1)
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=80)
x,y,yprime = EulerCromerMethod(.25)
plt.plot(x,Error1,label="EulerCromer Method,step=.25")
x,y,yprime = EulerCromerMethod(.02)
plt.plot(x,Error2,label="EulerCromer Method,step=.02")
x,y,yprime = MidpointMethod(.25)
plt.plot(x,Error3,label="Midpoint Method,step=.25")
x,y,yprime = MidpointMethod(.02)
plt.plot(x,Error4,label="Midpoint Method,step=.02")
plt.xlim(0,1.0)
plt.xlabel("X",fontsize=20)
plt.ylabel("Error",fontsize=20)
plt.legend(loc="upper left")
```

The most accurate result occurs with a step of .02 for the Euler Cromer method, as well as .02 for the Midpoint Method.

We can use numerical solvers and widgets to explore the dynamics of a projectile that is subject to a drag force proportional to the square of its velocity - "quadratic drag".

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def PPderiv(y,t,Vterr=44):
    
    g = 9.8
    m = 1 #kg
    Vx = y[0]
    Vy = y[1]
    x = y[2]
    y = y[3]
    
    dVxdt = (-g)*((Vx/(Vterr)**2))*np.sqrt((Vx)**2+(Vy)**2)
    dVydt = (-g)*(1+((Vy/(Vterr)**2)*np.sqrt((Vx)**2+(Vy)**2)))
    
    return np.array([ dVxdt, dVydt, Vx, Vy ])

t = np.arange(0.,200.+0.1,0.1)

#initial conditions
Vx0 = 19.28
Vy0 = 22.98
x0 = 0
y0 = 0

#pass the function deriv to the solver with initial values and integration variable
result = odeint(PPderiv,np.array([Vx0,Vy0,x0,y0]),t)


X = result[:,2]
Y = result[:,3]

plt.plot(X,Y,label="Trajectory")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0,100)
plt.ylim(0,100)
plt.legend(loc="best")
plt.show()
```

(b) Now create an interact object that uses sliders for the initial speed, terminal velocity, launch angle, and time. Have it plot two curves - one representing the solution with quadratic drag and the other in the absence of drag:

```python
from IPython.html.widgets import interact, interactive

def PPderiv2(y,t,Vterr=44):
    
    g = 9.8
    m = 1 #kg
    Vx = y[0]
    Vy = y[1]
    x = y[2]
    y = y[3]
    
    dVxdt = 0
    dVydt = (-g)
    
    return np.array([ dVxdt, dVydt, Vx, Vy ])


def Trajectory(theta=(pi/4),Vinit=30.,t=10,Vterr=40):
    t = np.arange(0.,t+0.1,0.1)
    Vx0=Vinit * np.cos(theta)
    Vy0=Vinit * np.sin(theta)

    #initial conditions
    x0 = 0
    y0 = 0

    #pass the function deriv to the solver with initial values and integration variable
    result = odeint(PPderiv,np.array([Vx0,Vy0,x0,y0]),t,args=(Vterr,))
    result2 = odeint(PPderiv2,np.array([Vx0,Vy0,x0,y0]),t,args=(Vterr,))


    X = result[:,2]
    Y = result[:,3]
    
    X2 = result2[:,2]
    Y2 = result2[:,3]
        
    plt.plot(X,Y,label="Trajectory")
    plt.plot(X2,Y2,label="Trajectory (w/o Drag)")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.xlim(0,100)
    plt.ylim(0,100)
    plt.legend(loc="best")
    plt.show()

v4 = interact(Trajectory,theta=(0.,(pi/2)),Vinit=(5.,50.),t=(.1,10.),Vterr=(20,120))
```
